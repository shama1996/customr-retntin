```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(df,col='2 How old are you? ',title='Distribution of Gender with age',hue='1Gender of respondent')
```

```python
uniplot(df,col='8 Which device do you use to access the online shopping?',title='Distribution of Gender with accessing device',hue='1Gender of respondent')
```

```python
uniplot(df,col='Which of the Indian online retailer would you recommend to a friend?',title='Distribution of Gender with online retailer',hue='1Gender of respondent')
```

```python
uniplot(df,col='Easy to use website or application',title='Distribution of easy website with speed delivery',hue='Speedy order delivery ')
```
